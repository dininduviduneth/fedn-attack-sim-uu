```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open("../metrics_trail.json", 'r') as json_file:
    data = json.load(json_file)
```

```python
metrics = {
    "training_loss": [item['training_loss'] for item in data],
    "training_accuracy": [item['training_accuracy'] for item in data],
    "test_loss": [item['test_loss'] for item in data],
    "test_accuracy": [item['test_accuracy'] for item in data]
}
```

```python
metric_type = "training_accuracy"
x = range(len(metrics[metric_type]))

# Plot actual data
plt.plot(x, metrics[metric_type], color='blue', linestyle='--')
    
# Add axis labels and title
plt.title(f'{metric_type}')
plt.xlabel(f'# of Rounds')
plt.ylabel(f'{metric_type}')

plt.show()
```
